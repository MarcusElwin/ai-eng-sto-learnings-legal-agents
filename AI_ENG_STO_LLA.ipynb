<div style="background-color: #2c3033; color: white; padding: 25px; border-radius: 0 0 10px 10px; margin-bottom: 30px;">
  <p style="margin-bottom: 20px; line-height: 1.6;">
    After working with legal tech, LLMs and now Agents for the past 18 months I have seen some learnings.
    Some of these are not unique to me or my experiences. Anyhow, some learnings in no inherent ordering:
  </p>

  <div style="margin: 30px 0;">
    <div style="display: flex; align-items: center; margin-bottom: 20px;">
      <span style="background-color: #444; border-radius: 50%; min-width: 40px; height: 40px; display: inline-flex; justify-content: center; align-items: center; margin-right: 20px; color: #ffd43b; font-weight: bold;">1</span>
      <span style="flex: 1;">
        Even if Agents are cool, they don't solve all problems<br>
        <small style="color: #ccc;">
          Agents excel at orchestrating tasks but may not outperform specialized models, or stacked chains of LLM calls or rule‚Äëbased systems for narrow jobs or tasks.
        </small>
      </span>
    </div>

  <div style="display: flex; align-items: center; margin-bottom: 20px;">
    <span style="background-color: #444; border-radius: 50%; min-width: 40px; height: 40px; display: inline-flex; justify-content: center; align-items: center; margin-right: 20px; color: #ffd43b; font-weight: bold;">2</span>
    <span style="flex: 1;">
      Choose what framework or library solves your <b>Use Case the best</b><br>
      <small style="color: #ccc;">
        Evaluate trade‚Äëoffs in features, integrations, and community support before committing to one solution.
      </small>
    </span>
  </div>

  <div style="display: flex; align-items: center; margin-bottom: 20px;">
    <span style="background-color: #444; border-radius: 50%; min-width: 40px; height: 40px; display: inline-flex; justify-content: center; align-items: center; margin-right: 20px; color: #ffd43b; font-weight: bold;">3</span>
    <span style="flex: 1;">
      Be able to quickly iterate on different models for different parts of your flow<br>
      <small style="color: #ccc;">
        Swap models in and out to optimize for cost, latency, and accuracy at each step of your agent pipeline.
      </small>
    </span>
  </div>

  <div style="display: flex; align-items: center; margin-bottom: 20px;">
    <span style="background-color: #444; border-radius: 50%; min-width: 40px; height: 40px; display: inline-flex; justify-content: center; align-items: center; margin-right: 20px; color: #ffd43b; font-weight: bold;">4</span>
    <span style="flex: 1;">
      Start Simple i.e. 1 agent or agentic flow and extend when necessary<br>
      <small style="color: #ccc;">
        Begin with a minimal proof‚Äëof‚Äëconcept and add complexity only when you‚Äôve validated real user value.
      </small>
    </span>
  </div>

  <div style="display: flex; align-items: center; margin-bottom: 20px;">
    <span style="background-color: #444; border-radius: 50%; min-width: 40px; height: 40px; display: inline-flex; justify-content: center; align-items: center; margin-right: 20px; color: #ffd43b; font-weight: bold;">5</span>
    <span style="flex: 1;">
      Use a tracing solution, you will need it<br>
      <small style="color: #ccc;">
        Capture each prompt, tool call, and response to debug issues and maintain an audit trail.
      </small>
    </span>
  </div>

  <div style="display: flex; align-items: center; margin-bottom: 20px;">
    <span style="background-color: #444; border-radius: 50%; min-width: 40px; height: 40px; display: inline-flex; justify-content: center; align-items: center; margin-right: 20px; color: #ffd43b; font-weight: bold;">6</span>
    <span style="flex: 1;">
      Make sure to track costs, Agent loops can be expensive<br>
      <small style="color: #ccc;">
        Monitor API usage and loop iterations to prevent unexpected billing spikes.
      </small>
    </span>
  </div>

  <div style="display: flex; align-items: center; margin-bottom: 20px;">
    <span style="background-color: #444; border-radius: 50%; min-width: 40px; height: 40px; display: inline-flex; justify-content: center; align-items: center; margin-right: 20px; color: #ffd43b; font-weight: bold;">7</span>
    <span style="flex: 1;">
      Give control to the end-user i.e. HIL or <i>human-in-the-loop</i><br>
      <small style="color: #ccc;">
        Allow humans to intervene, review, and override agent decisions at critical steps.
      </small>
    </span>
  </div>

  <div style="display: flex; align-items: center; margin-bottom: 20px;">
    <span style="background-color: #444; border-radius: 50%; min-width: 40px; height: 40px; display: inline-flex; justify-content: center; align-items: center; margin-right: 20px; color: #ffd43b; font-weight: bold;">8</span>

  <div style="display: flex; align-items: center; margin-bottom: 20px;">
    <span style="background-color: #444; border-radius: 50%; min-width: 40px; height: 40px; display: inline-flex; justify-content: center; align-items: center; margin-right: 20px; color: #ffd43b; font-weight: bold;">9</span>

Finally, let's try to add some tools to help the agent here

````python
from dataclasses import dataclass
from typing import Optional

# Dependencies class for your agent
@dataclass
class ContractDeps:
    effective_date: Optional[str] = None  # The effective date of the contract
````

````python
from pydantic_ai import RunContext

# Add calculator tool
@openai_agent.tool
async def calculator(ctx: RunContext[ContractDeps], expression: str) -> float:
    """Calculate the result of a mathematical expression.
    
    Args:
        ctx: The context with dependencies
        expression: Mathematical expression to evaluate (e.g., "2 + 3 * 4")
    """
    # Safety check for allowed characters
    allowed_chars = set("0123456789+-*/() .")
    if not all(c in allowed_chars for c in expression):
        raise ValueError(f"Invalid characters in expression: {expression}")

    try:
        return eval(expression)
    except Exception as e:
        return f"Error: {str(e)}"
````

````python
import re
from datetime import datetime, timedelta

# Add date calculator tool
@openai_agent.tool
async def date_calculator(
    ctx: RunContext[ContractDeps], 
    operation: str,
    date_text: str,
    value: Optional[int] = None,
    unit: Optional[str] = None
) -> str:
    """Calculate, parse, or manipulate dates.
    
    Args:
        ctx: The context with dependencies
        operation: Operation to perform (parse, add, subtract)
        date_text: Text description of a date or date in ISO format
        value: Optional value for add/subtract operations
        unit: Optional unit for add/subtract operations (days, months, years)
    """
    # Get reference date from dependencies if available
    reference_date = ctx.deps.effective_date
    if reference_date is None:
        # Default to today if no effective date is provided
        reference_date = datetime.now().strftime("%Y-%m-%d")

    # Parse the reference date
    try:
        ref_date = datetime.strptime(reference_date, "%Y-%m-%d")
    except ValueError:
        return f"Error: Invalid reference date {reference_date}"

    # Handle different operations
    if operation.lower() == "parse":
        # Check if already in ISO format
        iso_pattern = r"^\d{4}-\d{2}-\d{2}$"
        if re.match(iso_pattern, date_text):
            return date_text

        # Try common date formats
        date_formats = ["%B %d, %Y", "%d %B %Y", "%B %d %Y", "%m/%d/%Y", "%d/%m/%Y"]
        for fmt in date_formats:
            try:
                date_obj = datetime.strptime(date_text, fmt)
                return date_obj.strftime("%Y-%m-%d")
            except ValueError:
                continue

        # Handle relative dates
        years_match = re.search(r"(\d+)\s+year", date_text.lower())
        months_match = re.search(r"(\d+)\s+month", date_text.lower())
        days_match = re.search(r"(\d+)\s+day", date_text.lower())

        if years_match or months_match or days_match:
            result_date = ref_date

            if years_match:
                years = int(years_match.group(1))
                result_date = result_date.replace(year=result_date.year + years)

            if months_match:
                months = int(months_match.group(1))
                new_month = result_date.month + months
                new_year = result_date.year + (new_month - 1) // 12
                new_month = ((new_month - 1) % 12) + 1
                result_date = result_date.replace(year=new_year, month=new_month)

            if days_match:
                days = int(days_match.group(1))
                result_date += timedelta(days=days)

            return result_date.strftime("%Y-%m-%d")

        # If we can't parse it, return the original
        return date_text

    elif operation.lower() in ["add", "subtract"]:
        # First parse the date_text if not already in ISO format
        if not re.match(iso_pattern, date_text):
            for fmt in ["%B %d, %Y", "%d %B %Y", "%B %d %Y", "%m/%d/%Y", "%d/%m/%Y"]:
                try:
                    date_obj = datetime.strptime(date_text, fmt)
                    date_text = date_obj.strftime("%Y-%m-%d")
                    break
                except ValueError:
                    continue

        # If we have a date in ISO format, do the calculation
        if re.match(iso_pattern, date_text):
            try:
                date_obj = datetime.strptime(date_text, "%Y-%m-%d")
            except ValueError:
                return f"Error: Invalid date format {date_text}"
        else:
            # Use reference date if date_text couldn't be parsed
            date_obj = ref_date

        # Check required parameters
        if value is None or unit is None:
            return "Error: Both value and unit are required for add/subtract operations"

        # Perform the calculation
        if operation.lower() == "add":
            if unit.lower() in ["day", "days"]:
                result_date = date_obj + timedelta(days=value)
            elif unit.lower() in ["month", "months"]:
                new_month = date_obj.month + value
                new_year = date_obj.year + (new_month - 1) // 12
                new_month = ((new_month - 1) % 12) + 1
                result_date = date_obj.replace(year=new_year, month=new_month)
            elif unit.lower() in ["year", "years"]:
                result_date = date_obj.replace(year=date_obj.year + value)
            else:
                return f"Error: Unknown unit {unit}"
        else:  # subtract
            if unit.lower() in ["day", "days"]:
                result_date = date_obj - timedelta(days=value)
            elif unit.lower() in ["month", "months"]:
                new_month = date_obj.month - value
                new_year = date_obj.year
                while new_month <= 0:
                    new_month += 12
                    new_year -= 1
                result_date = date_obj.replace(year=new_year, month=new_month)
            elif unit.lower() in ["year", "years"]:
                result_date = date_obj.replace(year=date_obj.year - value)
            else:
                return f"Error: Unknown unit {unit}"

        return result_date.strftime("%Y-%m-%d")

    else:
        return f"Error: Unknown operation {operation}"
````
